```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris

iris = load_iris()

data = iris.data
target = iris.target
features = list(iris.feature_names)
names = [n.capitalize() for n in iris.target_names]

print(f'Dataset contains {data.shape[0]} entries of {data.shape[1]} features.')
print(f'Entries are classified in {len(names)} classes.')
```

## 5. Classification with `scatter`

```python
plt.figure()
xfeat = 0
yfeat = 1
scatter = plt.scatter(data[:, xfeat], data[:, yfeat], c=target)
plt.xlabel(features[xfeat])
plt.ylabel(features[yfeat])
plt.legend(scatter.legend_elements()[0], names)
plt.show()
```

```python
f = lambda x, y: np.sin(x) + np.cos(x + y)
X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(X, Y)

Z = f(X, Y)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
```
