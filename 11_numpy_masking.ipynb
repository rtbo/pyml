```python
from scipy import datasets
import matplotlib.pyplot as plt
```

```python
def process(img):
    (h, w) = img.shape
    img = img[h//4:-h//4, w//4:-w//4]
    img[img > 200] = 255
    img[img < 50] = 0
    return img
```

```python
face = datasets.face(gray=True)
print(face.shape)
plt.imshow(face, cmap='gray')
```

```python
zoom = process(face)
print(zoom.shape)
plt.imshow(zoom, cmap=plt.cm.gray)
```
