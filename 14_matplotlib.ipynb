```python
import numpy as np
```

```python
x = np.linspace(0, 2, 100)
Y = x**2
```

## Pyplot

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x, Y)
```

```python
plt.plot(x, Y, c="red", lw=3, ls="dashed")
```

## lifecycle

Figures can be customized between `plt.figure()` and `plt.show()`

```python
plt.figure(figsize=(10, 5))
plt.plot(x, Y)
plt.plot(x, x**3, c="red")
plt.title("Matplotlib")
plt.xlabel("x")
plt.ylabel("Y")
plt.grid()
plt.legend(["Y = x^2", "Y = x^3"])
plt.show()
```

### Subplots

```python
plt.figure(figsize=(5, 8))

plt.subplot(2, 1, 1)
plt.plot(x, Y)
plt.title("plot 1")

plt.subplot(2, 1, 2)
plt.plot(x, Y, c="r")
plt.title("plot 2")

plt.show()
```

## Object oriented

```python
fig, [ax1 , ax2] = plt.subplots(2, 1, sharex=True)
ax1.plot(x, Y)
ax2.set_title("empty")
plt.show()
```

## Exercice

Create a function that plots the result of experiences in a dictionary

```python
dataset = { f"Experiment {i+1}": np.random.randn(100) for i in range(4)}
```

```python
def plot_dataset(dataset):
    n = len(dataset)
    plt.figure(figsize=(12, 4*n))
    for k, i in zip(dataset.keys(), range(1, n+1)):
        plt.subplot(n, 1, i)
        plt.plot(dataset[k])
        plt.title(k)
    
    plt.show()
```

```python
plot_dataset(dataset)
```
